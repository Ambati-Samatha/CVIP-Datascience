# importing libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# getting datasets 

```python
unemployment = pd.read_csv('Unemployment in India.csv')          #Unemployment data of india
UnEmp_covid = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')  #Unemployment data during covid
```

```python
region_data = unemployment.groupby('Region').sum()

fig, ax = plt.subplots(figsize=(20, 8))
ax.pie(region_data[' Estimated Unemployment Rate (%)'], labels=region_data.index, autopct='%1.1f%%', startangle=90,textprops={'fontsize':8})
ax.legend(loc='upper right', labels=[f'{l}: {s}' for l, s in zip(region_data, region_data[' Estimated Unemployment Rate (%)'])], prop={'size': 5})
ax.set_title('Employment Distribution by Region')
plt.show()
```

```python
region_data1 = UnEmp_covid.groupby('Region').sum()

fig, ax = plt.subplots(figsize=(20, 8))
ax.pie(region_data1[' Estimated Unemployment Rate (%)'], labels=region_data1.index, autopct='%1.1f%%', startangle=90,textprops={'fontsize':8})
ax.legend(loc='upper right', labels=[f'{l}: {s}' for l, s in zip(region_data1, region_data1[' Estimated Unemployment Rate (%)'])], prop={'size': 5})
ax.set_title('Employment Distribution by Region')
plt.show()
```

```python
x_regions = np.array(unemployment['Region'])
x_regions_cov = np.array(UnEmp_covid['Region'])
y_unp = np.array(unemployment[' Estimated Labour Participation Rate (%)'])
y_unp_covid = np.array(UnEmp_covid[' Estimated Labour Participation Rate (%)'])
```

```python
fig = plt.figure(figsize=(20, 10))

plt.bar(x_regions, y_unp, color= 'orange')
plt.plot(x_regions_cov, y_unp_covid, color= 'blue')
plt.ylim(0, 100)

labels = sorted(set(unemployment['Region']))
plt.xticks(labels, rotation='vertical')

plt.xlabel('Region by States')
plt.ylabel('Estimated Labour participation  rate')
plt.title('Estimated Labour Participation rate of 2019 vs 2020')

plt.show()
```

```python
# Select the numerical columns for correlation
numerical_columns = [' Estimated Unemployment Rate (%)', ' Estimated Employed',
                     ' Estimated Labour Participation Rate (%)']

# Calculate the correlation matrix
correlation_matrix = unemployment[numerical_columns].corr()

# Create the heatmap
plt.figure(figsize=(8, 5.5))
sns.set_context('notebook', font_scale=1)
sns.heatmap(correlation_matrix, annot=True, cmap="cool")
plt.title('Correlation Heatmap of Unemployment Data')
plt.show()
```

```python
# Group data by region and date
grouped_df = unemployment.groupby(["Region", " Date"]).sum()

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=" Date", y=" Estimated Unemployment Rate (%)", hue="Region", data=grouped_df)
plt.title("Estimated Unemployment Rate Over Time by Region")
plt.xlabel("Date")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Group data by region and date
grouped_df1 = UnEmp_covid.groupby(["Region", " Date"]).sum().reset_index()

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=" Date", y=" Estimated Unemployment Rate (%)", hue="Region", data=grouped_df1)
plt.title("Estimated Unemployment Rate Over Time by Region in 2020")
plt.xlabel("Date")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Create the violin plot
plt.figure(figsize=(15, 8))
sns.violinplot(x='Region', y=' Estimated Unemployment Rate (%)', data=UnEmp_covid)
plt.title('Estimated Unemployment Rate in 2020 by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```
